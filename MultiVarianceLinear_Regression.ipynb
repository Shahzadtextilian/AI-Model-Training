```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df=pd.read_csv("/content/drive/MyDrive/AI/Student_Performance.csv")
df.rename(columns={'Extracurricular Activities':'Extracurricular_Activities','Hours Studied':'Hours_Studied','Previous Scores':'Previous_Scores', 'Sleep Hours':'Sleep_Hours','Sample Question Papers Practiced':'Papers_Practiced','Performance Index':'Performance_Index'},inplace=True)
df.head(10)
```

```python
#df = df.drop(columns=['Extracurricular Activities'])
df.drop(columns=['Extracurricular_Activities'], inplace=True)
df
```

```python
lr=linear_model.LinearRegression()
lr.fit(df[['Hours_Studied','Previous_Scores','Sleep_Hours','Papers_Practiced']],df.Performance_Index)
```

```python
#test the model
lr.predict([[2,40,6,0]])
```

```python
lr.score(df[['Hours_Studied','Previous_Scores','Sleep_Hours','Papers_Practiced']],df.Performance_Index)
lr.coef_
```

```python
lr.intercept_
```
